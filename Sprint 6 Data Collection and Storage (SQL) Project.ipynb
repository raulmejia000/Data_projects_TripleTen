```python
# Load the dataset
# dataset7 = Data on rides from the Loop to O'Hare International Airport
dataset7 = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
# converting column names to lower case
dataset7.columns = dataset7.columns.str.lower()
```

```python
# show the data
print(dataset7.head())
```

```python
print(dataset7.tail())
```

```python
# check the datatypes
# get summary of datatypes
# start_ts = pickup date and time
# weather_conditions = weather conditions at the moment the ride started
# duration_seconds = ride duration in seconds

print(dataset7.info())
```

```python
# get basic stats

print(dataset7.describe())
```

```python
# checking for missing values

print(dataset7.isnull().sum())
```

```python
# check for duplicates

print(dataset7.duplicated().sum())
```

```python
# remove duplicates

dataset7 = dataset7.drop_duplicates()

# check again to make sure

print(dataset7.duplicated().sum())
```

```python
# Confirm all rides are from Saturdays
dataset7['start_ts'] = pd.to_datetime(dataset7['start_ts'])
dataset7['day_of_week'] = dataset7['start_ts'].dt.day_name()

# Check the unique days
print('Days present in dataset:', dataset7['day_of_week'].unique())
```

```python
# Split the data
rainy_rides = dataset7[dataset7['weather_conditions'] == 'Bad']['duration_seconds']
good_rides = dataset7[dataset7['weather_conditions'] == 'Good']['duration_seconds']

# Perform the t-test
results = stats.ttest_ind(rainy_rides, good_rides, equal_var=False)  # Welch's t-test (safer)

# Display results
print('t-statistic:', results.statistic)
print('p-value:', results.pvalue)

# Conclusion
if results.pvalue < 0.05:
    print("We reject the null hypothesis: Average ride duration differs on rainy Saturdays.")
else:
    print("We fail to reject the null hypothesis: No significant difference in average ride duration.")
```

I tested the hypothesis that rainy weather affects the average duration of taxi rides from the Loop to O'Hare International Airport on Saturdays. The null hypothesis assumed no difference in average ride duration between rainy and non-rainy conditions, while the alternative hypothesis suggested a difference. I used an independent two-sample t-test, appropriate for comparing two independent groups with continuous numerical data. The significance level (alpha) was set at 0.05, A common standard that balances the risk of making a Type I error with maintaining reasonable sensitivity.. Based on the p-value obtained, we determined whether to reject or fail to reject the null hypothesis.
