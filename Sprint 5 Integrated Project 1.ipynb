```python
# Define relevant period: Focus on last 5 years (2012-2016)
relevant_years = df[(df['year_of_release'] >= 2012) & (df['year_of_release'] <= 2016)]

# Group by platform and year, then sum sales
platform_sales = relevant_years.groupby(["platform", "year_of_release"])["total_sales"].sum().reset_index()

# Identify top platforms based on total sales in recent years
top_platforms = platform_sales.groupby("platform")["total_sales"].sum().sort_values(ascending=False).head(8).index

# Filter only the top-performing platforms
platform_sales_recent = platform_sales[platform_sales["platform"].isin(top_platforms)]

# Pivot the table for easier plotting
platform_pivot = platform_sales_recent.pivot(index="year_of_release", columns="platform", values="total_sales").fillna(0)

# Plot the filtered sales trends
plt.figure(figsize=(12, 6))
for platform in platform_pivot.columns:
    plt.plot(platform_pivot.index, platform_pivot[platform], label=platform)

plt.xlabel("Year of Release")
plt.ylabel("Total Sales (Millions)")
plt.title("Sales Trends for Top Platforms (2012-2016)")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Select two platforms for comparison (PS4 and XOne)
platforms = ['PS4', 'XOne']
df_filtered = df[df['platform'].isin(platforms)][['platform', 'critic_score', 'user_score', 'total_sales']].dropna()

# Create scatter plots for both platforms
plt.figure(figsize=(12, 10))

for i, platform in enumerate(platforms):
    plt.subplot(2, 2, i+1)
    sns.scatterplot(data=df_filtered[df_filtered['platform'] == platform], x='critic_score', y='total_sales', alpha=0.5)
    plt.title(f"Critic Score vs Total Sales ({platform})")
    plt.xlabel("Critic Score")
    plt.ylabel("Total Sales (Millions)")

    plt.subplot(2, 2, i+3)
    sns.scatterplot(data=df_filtered[df_filtered['platform'] == platform], x='user_score', y='total_sales', alpha=0.5)
    plt.title(f"User Score vs Total Sales ({platform})")
    plt.xlabel("User Score")
    plt.ylabel("Total Sales (Millions)")

plt.tight_layout()
plt.show()

# Compute correlation coefficients for both platforms
for platform in platforms:
    platform_data = df_filtered[df_filtered['platform'] == platform]
    
    critic_corr, _ = pearsonr(platform_data['critic_score'], platform_data['total_sales'])
    user_corr, _ = pearsonr(platform_data['user_score'], platform_data['total_sales'])
    
    print(f"Platform: {platform}")
    print(f"Correlation between Critic Score and Total Sales: {critic_corr:.2f}")
    print(f"Correlation between User Score and Total Sales: {user_corr:.2f}")
    print("-" * 40)
```

The analysis shows a moderate positive correlation (0.41) between critic scores and total sales for both PS4 and XOne, suggesting that higher critic ratings may influence game sales on these platforms. However, user scores show little to no correlation with sales (-0.03 for PS4, -0.09 for XOne), indicating that user ratings do not significantly impact a game’s commercial success. The scatter plots support this, showing a clearer trend between critic scores and sales, while user scores appear more scattered. This suggests that critics’ reviews may play a role in influencing purchasing decisions, whereas user ratings are less reliable, potentially due to personal biases or varied expectations.

```python
from scipy.stats import ttest_ind

# Filter data to match the 2012-2016 time interval
df_filtered = df[(df['year_of_release'] >= 2012) & (df['year_of_release'] <= 2016)]

# ---------------- Hypothesis 1: Xbox One vs. PC User Ratings ----------------
# Define Hypotheses:
# H₀: There is no significant difference in average user ratings between Xbox One and PC games.
# H₁: There is a significant difference in average user ratings between Xbox One and PC games.

# Extract user ratings for Xbox One and PC
xbox_one_ratings = df_filtered[df_filtered['platform'] == 'XOne']['user_score'].dropna()
pc_ratings = df_filtered[df_filtered['platform'] == 'PC']['user_score'].dropna()

# Perform t-test
t_stat_xbox_pc, p_value_xbox_pc = ttest_ind(xbox_one_ratings, pc_ratings, equal_var=False)

# ---------------- Hypothesis 2: Action vs. Sports Genre User Ratings ----------------
# Define Hypotheses:
# H₀: There is no significant difference in average user ratings between Action and Sports games.
# H₁: There is a significant difference in average user ratings between Action and Sports games.

# Extract user ratings for Action and Sports genres
action_ratings = df_filtered[df_filtered['genre'] == 'Action']['user_score'].dropna()
sports_ratings = df_filtered[df_filtered['genre'] == 'Sports']['user_score'].dropna()

# Perform t-test
t_stat_action_sports, p_value_action_sports = ttest_ind(action_ratings, sports_ratings, equal_var=False)

# ---------------- Output Results ----------------
alpha = 0.05  # Significance Level

print(f"Xbox One vs. PC: t-statistic = {t_stat_xbox_pc}, p-value = {p_value_xbox_pc}")
if p_value_xbox_pc < alpha:
    print("Reject the null hypothesis: There is a significant difference in user ratings between Xbox One and PC games.")
else:
    print("Fail to reject the null hypothesis: No significant difference in user ratings between Xbox One and PC games.")

print(f"Action vs. Sports: t-statistic = {t_stat_action_sports}, p-value = {p_value_action_sports}")
if p_value_action_sports < alpha:
    print("Reject the null hypothesis: There is a significant difference in user ratings between Action and Sports games.")
else:
    print("Fail to reject the null hypothesis: No significant difference in user ratings between Action and Sports games.")

# ---------------- Explanation of the test and significance level ----------------
print("\nStatistical Test Used: Independent t-test")
print("The independent t-test was chosen because we are comparing the means of two independent groups for each hypothesis.")
print("Alpha Level (α): Set at 0.05 to ensure a 95% confidence level, balancing the risk of Type I errors while maintaining statistical power.")
```
